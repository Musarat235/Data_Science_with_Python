```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.barplot(data=df, x='Gender', y='Age (years)-Float/Int', palette='flare',hue='Qualification_completed')
plt.show()
```

```python
sns.catplot(data=df, x='How many hours you code a day? (int) e.g: 5,4,3', y='What are you?', kind = 'bar', palette='pastel',hue='Prepaid or Postpaid')
plt.show()
```

```python
sns.lineplot(data=df, x='Where do you live?', y='Light kitni der band hti hy? int', palette='dark',hue='Location')
plt.show()
```

```python
sns.boxenplot(data=df, x='Your Weight in kg? (float)', y='Your favorite programming language?')
plt.show()
```

```python
sns.barplot(data=df, x='Your Weight in kg? (float)', y='Your favorite programming language?',
            linewidth=2.5, facecolor='lightblue',
            errcolor=".2", edgecolor="black")
plt.show()
```

```python
sns.boxplot(data=df, x='Gender', y='Height in cm? Freelancer- (Float)',hue='Where do you live?',dodge=True, palette='Set2')
plt.show()
```

```python
sns.boxplot(x=df['Your Weight in kg? (float)'])
plt.show()
```

```python
# Set up a grid of axes with a polar projection
g = sns.FacetGrid(df, col="Where do you live?", hue="Where do you live?",
                  subplot_kws=dict(projection='polar'), height=4.5,
                  sharex=False, sharey=False, despine=False)

# Draw a scatterplot onto each axes in the grid
g.map(sns.scatterplot, "Age (years)-Float/Int", "What are you?")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

# Load the example diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
```

```python
sns.set_theme(style="darkgrid")
sns.displot(
    df, x="Age (years)-Float/Int", col="Marital Status?", row="Gender",
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),
)
```

```python
sns.pairplot(data=df, hue='Prepaid or Postpaid', palette='rocket_r')
```

```python
# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(
    data=df,
    x="How many hours you code a day? (int) e.g: 5,4,3", y="Light kitni der band hti hy? int", hue="Marital Status?",
    height=5
)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Hour you code a day", "Light kitni der band hti hy? int")
```
