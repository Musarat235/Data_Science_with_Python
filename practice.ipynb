```python
# Set up a grid of axes with a polar projection
g = sns.FacetGrid(df, col="Where do you live?", hue="Where do you live?",
                  subplot_kws=dict(projection='polar'), height=4.5,
                  sharex=False, sharey=False, despine=False)

# Draw a scatterplot onto each axes in the grid
g.map(sns.scatterplot, "Age (years)-Float/Int", "What are you?")
```

```python
sns.set_theme(style="darkgrid")
sns.displot(
    df, x="Age (years)-Float/Int", col="Marital Status?", row="Gender",
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),
)
```

```python
sns.pairplot(data=df, hue='Prepaid or Postpaid', palette='rocket_r')
```

```python
g = sns.lmplot(
    data=df,
    x="How many hours you code a day? (int) e.g: 5,4,3", y="Light kitni der band hti hy? int", hue="Marital Status?",
    height=5
)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Hour you code a day", "Light kitni der band hti hy? int")
```

# Plotly 3D Graph practice

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')
z = z_data.values
sh_0, sh_1 = z.shape
x, y = np.linspace(0, 1, sh_0), np.linspace(0, 1, sh_1)
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

```python
import plotly.express as px

df = px.data.tips()

fig = px.box(df, x="day", y="total_bill", color="smoker")
fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear" by default
fig.show()
```

```python
import plotly.figure_factory as ff
import numpy as np

# Add histogram data
x1 = np.random.randn(200) - 2
x2 = np.random.randn(200)
x3 = np.random.randn(200) + 2
x4 = np.random.randn(200) + 4

# Group data together
hist_data = [x1, x2, x3, x4]

group_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2)
fig.show()
```
